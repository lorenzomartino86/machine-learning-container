### Time Reseampling

```python
df = pd.read_csv('time_data/walmart_stock.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.resample(rule='A')
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='Q').mean()
```

```python
df.resample(rule='A').max()
```

```python
def first_day(entry):
    return entry[0]
df.resample(rule='A').apply(first_day)
```

```python
df['Close'].resample('Q').mean().plot(kind='bar', figsize=(16,6))
```

#### Rolling and expanding

```python
# With a legend automatically attached by matplotlib:
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16,6))
```

```python
# Expanding
df['Close'].expanding().mean().plot(figsize=(16,6))
```

### Bollinger Bands Generation
Above and below a moving average.

```python
# Closing price 20 Days MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper = 20 Days MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lover = 20 Days MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6))
```

```python
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].tail(100).plot(figsize=(16,6))
```
